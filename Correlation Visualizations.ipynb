# NBS Correlation Bar Charts

```
corr_nbs = pd.DataFrame()

demo_factor_lbl = ['MINORPCT', 'LOWINCPCT', 'LESSHSPCT', 'LINGISOPCT', 'OVER64PCT']
demo_factor = ['% People of color', '% Low income', '% Less than HS Education', '% Limited English speaking households', '% Over age 64']
cuv = [-0.4327322799236715, -0.26575474840509455, -0.282273657118961, -0.27449267144713296, 0.32549225668988263]

corr_nbs['Demographic Factor LBL'] = demo_factor_lbl
corr_nbs['Carbon Urbanforest Value'] = cuv
corr_nbs['Demographic Factor'] = demo_factor

corr_nbs
```

```
import matplotlib.pyplot as plt
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.barh(demo_factor, cuv, color='darkolivegreen')
#plt.title('Correlation between demographic factors and carbon urbanforest value')
plt.ylabel('Demographic Factor')
plt.xlabel('Carbon Urbanforest Value')
plt.show()
```
